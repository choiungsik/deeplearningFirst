## X데이터 생성 (공부한시간, 집과의 거리 - 성적)

```python
x1 = np.random.uniform(low =0.0, high=1.0, size=100).reshape(-1, 1) # 100행 1열 데이터생성
x2 = np.random.uniform(low =0.0, high=1.0, size=100).reshape(-1, 1)
```

```python
x1.shape
```

```python
X = np.concatenate((x1, x2), axis=1)
X.shape
```

#### y = w1*x1 + w2*x2 + b

```python
y = (12*x1) + (-5*x2) + 3
y = y.reshape(-1)
```

```python
y
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

#### 객체생성

```python
model = Sequential()
```

```python
# 신경망 설계
model.add(Dense(1, input_dim=2))  # 1개의 데이터를 뽑아내기 input_dim : 입력차원수
model.add(Dense(4)) 
model.add(Dense(2)) 
# 모델에 Densc를 여러개 넣을 수 있고 첫번째 이후로는 input_dim을 설정하지 않아도 됨
```

```python
# 신경망 구축
model.compile(loss ='mse', optimizer='SGD')  # mse : 평균제곱오차, SGD : 경사하강법을 이용한 최적화
```

```python
# 학습
model.fit(X, y, epochs=4000)
```

```python
model.get_layer('dense').get_weights()
```
