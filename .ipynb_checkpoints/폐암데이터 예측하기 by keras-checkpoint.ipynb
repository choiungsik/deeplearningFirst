## 폐암데이터 예측하기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
Data_set = np.loadtxt('./data/ThoraricSurgery.csv', delimiter=',')
```

```python
Data_set
```

```python
X =Data_set[:,0:17]
y= Data_set[:,17]
```

```python
X.shape
```

```python
model =Sequential()
model.add( Dense(30, input_dim=17, activation='relu'))
model.add( Dense(1, activation='sigmoid'))
```

mean_squared_error
binary_crossentropy

```python
model.compile( loss ='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=1000, batch_size=10, verbose=1)
# batch_size : 1회학습당 데이터의 크기
# verbose : 학습진행 표시여부 1이 기본출력, 0은 생략, 2는 OneLinePerpoch
```

```python
model.predict(X)
```

```python
model.evaluate(X, y)
```
